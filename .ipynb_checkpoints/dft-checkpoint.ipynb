```python
import numpy as np
import json
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
x = np.arange(0, 4, 10/1000)

p = np.pi / 6
y = np.sin(np.pi * (x + p)) + (np.sin(np.pi * 3 * (x + p)) * 1/3) + (np.sin(np.pi * 5 * (x + p)) * 1/5)

plt.plot(x, y)
plt.show()
```

```python
def find_freq(time, sig, freq):
    s = np.sin(np.pi * freq * time)
    c = np.cos(np.pi * freq * time)
    
    re = sig * c
    im = sig * s
    
    amp = 2 * np.sqrt(re.mean() ** 2 + im.mean() ** 2)
    phase = np.arctan((im.mean() / re.mean()))
    
    return (amp, phase)
```

```python
r = {}
for f in range(1, 8):
    (amp, phase) = find_freq(x, y, f)
    amp = 0 if amp < 0.000001 else amp
    r[f] = { 'amp': amp, 'phase': phase}
    
r
```

```python
sig = 0
for f in r:
    amp = r[f]['amp']
    ph1 = r[f]['phase']
    if amp != 0:
        print(f, amp, ph1)
        sig += np.sin(np.pi * f * x + ph1) * amp
#(np.sin(np.pi * 3 * x + 3 * p) / 3)
plt.plot(x, sig)
plt.show()
```
