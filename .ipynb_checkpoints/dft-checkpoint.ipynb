```python
import numpy as np
import json
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
x = np.arange(0, 4, 10/1000)

p = np.pi / 6
y = np.sin(np.pi * x + p) + (np.sin(np.pi * 3 * x + 3 * p) / 3) + (np.sin(np.pi * 5 * x + 5 * p) / 5)

plt.plot(x, y)
plt.show()
```

```python
def find_freq(time, sig, freq):
    s = np.sin(np.pi * freq * time)
    c = np.cos(np.pi * freq * time)
    
    re = sig * c
    im = sig * s
    
    amp = 2 * np.sqrt(re.mean() ** 2 + im.mean() ** 2)
    phase = np.arctan((re.mean() / im.mean()))
    
    return (amp, phase)
```

```python
r = {}
for f in range(1, 8):
    (amp, phase) = find_freq(x, y, f)
    amp = 0 if amp < 0.000001 else amp
    r[f] = { 'amp': amp, 'phase': phase}
    
r
```
