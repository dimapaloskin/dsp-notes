## Odd and Even Parts of Signal

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Definition

**Even**

$ f(t) = f(-t) $

**Odd**

$ f(t) = -f(-t) $

### Sinusoidal signal

```python
def compose_signal(freq, phase = 0, amp = 1, periods = 2):
    sine_amp = amp * np.sin(phase)
    cos_amp = amp * np.cos(phase)
    
    return lambda x: sine_amp * np.sin(2 * np.pi * x * freq) + cos_amp * np.cos(2 * np.pi * x * freq)
```

**Create original signal**

```python
periods = 1
freq = 1
amp = 0.5
phase = np.pi / 6
```

```python
signal = compose_signal(freq, phase, amp, periods)


x = np.arange(0, periods, 0.01)
y_original = signal(x)

plt.plot(x, y_original, color='#489cb5')

plt.title('Original signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.show()
```

**Find Odd and Even parts**

```python
y_even = 0.5 * (signal(x) + signal(-x))
y_odd = 0.5 * (signal(x) - signal(-x))

plt.plot(x, y_even, color='#489cb5')
plt.plot(x, y_odd, color='#db5670')

plt.title('Odd and Even parts of signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

**Compare orginal and summarized signals**

```python
y_sum = y_even + y_odd

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))

axes[0].plot(x, y_original)
axes[1].plot(x, y_sum, color='#db5670')

axes[0].set_title('Original Signal')
axes[1].set_title('Summarized signal')

for ax in axes:
    ax.set_xlabel('Time')
    ax.set_ylabel('Amplitude')

plt.show()
```
