## Play with Unit Step and Rectifier Functions

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Definition

**Unit Step Function**
$$ u(t) = \begin{cases}
    1 &  t\geq0\\
    0 & t < 0
  \end{cases}  $$

**Rectifier Function**

$$ r(t) = \begin{cases}
    t &  t\geq0\\
    0 & t < 0
  \end{cases}$$

### Define and plot

```python
# Unit Step
def u(bias = 0):
    fn = lambda x: 1 if x > -bias else 0
    return np.frompyfunc(fn, 1, 1)
# Rectifier
def r(bias = 0):
    fn = lambda x: x + bias if x > -bias else 0
    return np.frompyfunc(fn, 1, 1)
```

```python
x = np.arange(-5, 5, 0.01)
```

```python
y_u = u()(x)
y_r = r()(x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))

axes[0].plot(x, y_u)
axes[1].plot(x, y_r, color='#db5670')

axes[0].set_title('Unit Step Function')
axes[1].set_title('Rectifier Function')

for ax in axes:
    ax.set_ylim(-2, 2)
    ax.set_xlabel('Time')
    ax.set_ylabel('Value')

plt.show()
```

## Compose

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))

y_c1 = 2 * u(2)(x) + r(-1)(x) - 2 * r(-2)(x) + r(-3)(x) - 2 * u(-4)(x)
y_c2 = -2 * r(2)(x) + 4 * r(0)(x) - 2 * r(-2)(x) + 2 * u(-2)(x) -  2 * u(-4)(x)  

for ax in axes:
    ax.set_ylim(-5, 5)
    ax.set_xlabel('Time')
    ax.set_ylabel('Value')

axes[0].plot(x, y_c1)
axes[1].plot(x, y_c2)

plt.show()
```
